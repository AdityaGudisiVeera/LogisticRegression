```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
employee_df = pd.read_csv('employee_turnover.csv')
```

```python
#column names
employee_df.columns
```

```python
#Information about the dataset
employee_df.info()
```

```python
#Checking for missing values
employee_df.isna().sum()
```

```python
# Assuming 'Employee_Turnover' is the target variable and the rest are features
X = employee_df.drop('Employee_Turnover', axis=1)
Y = employee_df['Employee_Turnover']
```

```python
#Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#checking the distribution of target variable in training set
print(Y_train[Y_train == 1].count())
print(Y_train[Y_train == 0].count())
```

```python
#Training the Logistic Regression model
model = LogisticRegression()
model.fit(X_train, Y_train)

#Predicting the test set results
Y_pred = model.predict(X_test)
```

```python
#Accuracy of the model
accuracy = accuracy_score(Y_test, Y_pred)
print(f'Accuracy of the Logistic Regression model: {accuracy:.2f}')
```
